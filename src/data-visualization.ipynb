```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = gpd.read_file("../data/final_dataset_mest.geojson")

df['crime_density'] = df['crime_total'] / df['area_km2']
df['crime_attractor_density'] = (df['poi_pub'] + df['poi_nightclub'] + df['poi_bar']) / df['area_km2']
df['crime_generator_density'] = (df['poi_cafe'] + df['poi_museum'] + df['poi_attraction']) / df['area_km2']
df['tourist_density'] = df['nights_non_residents'] / df['area_km2']
df['foreigner_density'] = df['foreigners'] / df['area_km2']
df['university_density'] = df['people_university'] / df['area_km2']
df['airbnb_economy_density'] = df['airbnb_economy'] / df['area_km2']
df['airbnb_mid'] = (df['airbnb_count'] - (df['airbnb_economy'] + df['airbnb_luxury'])).clip(lower=0)
df['crime_Burglary_density'] = df['crime_Burglary'] / df['area_km2']
df['crime_Disorder_density'] = df['crime_Disorder'] / df['area_km2']
df['crime_Theft_density'] = df['crime_Theft'] / df['area_km2']
df['crime_Violent_density'] = df['crime_Violent'] / df['area_km2']
df['crime_Other_density'] = df['crime_Other'] / df['area_km2']

df = df.fillna(0.0)
```

### Spatial Distribution Maps and Histograms

#### Economy Airbnb Density

```python
# vzor

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))

df.plot(
    column='airbnb_economy_density', 
    ax=ax1, 
    legend=True, 
    legend_kwds={'label': "Units per $km^2$"},
    edgecolor='white',
    linewidth=0.5)
ax1.set_title('Economy Airbnb Density by District', fontsize=15, fontweight='bold')
ax1.axis('off')

ax2.hist(df['airbnb_economy_density'], bins=20, edgecolor='black', color='steelblue', alpha=0.7)
ax2.set_title('Economy Airbnb Density Distribution', fontsize=15, fontweight='bold')
ax2.set_xlabel('Economy Airbnb Density', fontsize=12)
ax2.set_ylabel('Frequency', fontsize=12)
ax2.axvline(df['airbnb_economy_density'].mean(), color='red', linestyle='--', linewidth=2, label='Mean')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
#plt.savefig('../figs/economy_airbnb_map_histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Scatter Plots

#### Crime Density vs Distance from Center & Mhd Density

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))

ax1.scatter(df['dist_center_km'], 
            df['crime_density'], 
            alpha=1, s=30, linewidth=0.5, color='darkgreen')
ax1.set_xlabel('Distance from Center (km)')
ax1.set_ylabel('Crime Density')
ax1.set_title('Crime vs Distance from Center', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)

z1 = np.polyfit(df['dist_center_km'], df['crime_density'], 2)
p1 = np.poly1d(z1)
x_smooth = np.linspace(df['dist_center_km'].min(), df['dist_center_km'].max(), 100)
ax1.plot(x_smooth, p1(x_smooth), "r--", linewidth=1)



ax2.scatter(df['mhd_density'], 
            df['crime_density'], 
            alpha=1, s=30, linewidth=0.5, color='darkgreen')
ax2.set_xlabel('MHD Density')
ax2.set_ylabel('Crime Density')
ax2.set_title('Crime vs Public Transport Density', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)

z2 = np.polyfit(df['mhd_density'], df['crime_density'], 1)
p2 = np.poly1d(z2)
ax2.plot(df['mhd_density'], 
         p2(df['mhd_density']), 
         "r--", linewidth=1)

plt.tight_layout()
plt.show()
```
